```python
import os
import datetime

import IPython
import IPython.display
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pathlib import Path 
```

# Функция построения единого графика

```python
def plot_pred(n):
    
    #Номер набора
    start_conting = 440
    #Количество точек в датафрейме
    t = 81
    #Входное окно
    input_t = 40
    
    print("______________")
    
    #Для примера считаем данные из тестовой выборки, т.е. эти данные не используются при обучении 
    
    #graph_a = graph[ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]
    
    #test_csv = df[t * (start_conting + n - 1): (start_conting + n) * t]
    
    time_input = np.array(range(0, t * 10 ,10))
    
    time_input_pred = np.array(range(input_t * 10 , (t - 2) * 10, 10))
    
    limit = np.full((t - input_t - 2 ), 180)
    
    graph_time = np.array(range((input_t - 13) * 10 , 10 * (60 + input_t - 13), 10))

    fig, axes = plt.subplots()

    fig.set_figwidth(13)
    fig.set_figheight(13)


    
    
    #Чита
    axes.plot(time_input,
              pd.read_csv('prediction/chita/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Чита)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/chita/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Читинской ТЭЦ - 1',
              linewidth = 1, alpha=0.8)
    axes.plot(time_input_pred, limit, linewidth = 3)
    axes.plot(graph_time,
              pd.read_csv('prediction/chita/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Чита)',
              linewidth = 2, alpha=0.9)
    
    #Белово
    axes.plot(time_input,
              pd.read_csv('prediction/belovo/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Белово)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/belovo/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Беловской ТЭЦ',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/belovo/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Белово)',
              linewidth = 2, alpha=0.9)

    #Берёза
    axes.plot(time_input,
              pd.read_csv('prediction/bereza/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Белово)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/bereza/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Берёзовской ГРЭС',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/bereza/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Берёза)',
              linewidth = 2, alpha=0.9)
    
    
    #Богучан
    axes.plot(time_input,
              pd.read_csv('prediction/boguchan/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Богучанская)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/boguchan/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Богучанская ГЭС',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/boguchan/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Богучанская)',
              linewidth = 2, alpha=0.9)
    
    
    
    #Братск 
    axes.plot(time_input,
              pd.read_csv('prediction/bratsk/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Братск)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/bratsk/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Братская ГЭС',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/bratsk/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Братск)',
              linewidth = 2, alpha=0.9)
    
    
    #Гусиноозерск 
    axes.plot(time_input,
              pd.read_csv('prediction/gusi/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Братск)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/gusi/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Гусиноозерской ГРЭС',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/gusi/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Гисноозерск)',
              linewidth = 2, alpha=0.9)
    
    
    
    #Иркутск 
    axes.plot(time_input,
              pd.read_csv('prediction/irkutsk/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Братск)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/irkutsk/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Иркутской ГЭС',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/irkutsk/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Иркутск)',
              linewidth = 2, alpha=0.9)
    
    
    
    
    #Красноярск 
    axes.plot(time_input,
              pd.read_csv('prediction/kras/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Братск)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/kras/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Красноярской ГЭС',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/kras/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Красноярск)',
              linewidth = 2, alpha=0.9)
    
        #Нижневартовск 
    axes.plot(time_input,
              pd.read_csv('prediction/vartovsk/df/df.csv')[t * (start_conting + n - 1): (start_conting + n) * t]['delta'],
              '--',
              #label = 'Фактическое значение до ликвидации возмущения (Братск)',
              linewidth = 2, alpha=0.8)
    axes.plot(time_input_pred,
              pd.read_csv(f'prediction/vartovsk/{n}.csv')['delta'][:39],
              'o-',
              label = f'Прогноз угла относительно Нижневартовская ГРЭС',
              linewidth = 1, alpha=0.8)
    axes.plot(graph_time,
              pd.read_csv('prediction/vartovsk/graph/graph_a.csv')
              [ 60 * (start_conting + n - 1) : (start_conting + n )  * 60]['delta'],
              '--',
              label = f'Фактическое значение после ликвидации возмущения {n + 1} (Нижневартовск)',
              linewidth = 2, alpha=0.9)
    
    
    
    
    
    plt.ylim(0,500)
    plt.xlim(0,(47 + input_t)*10)

    #Добавим оси
    plt.xlabel('t, мс',fontsize=13)
    plt.ylabel('Угол ротора, град',fontsize=13)
    plt.title(f'Возмущение номер {n + 1}')

    #Добавим решетку 
    axes.grid(which='major',
            color = 'k', 
            linewidth = 0.8)

    axes.grid(which='minor', 
            color = 'k', 
            linestyle = ':',
             linewidth = 0.6)
    axes.minorticks_on()
    axes.legend(fontsize=13 , bbox_to_anchor=(1, 0.6))

    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.show()
    print("______________")
    #fig.savefig('predict.png', dpi=300, facecolor='white')
```
